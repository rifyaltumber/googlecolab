```python
import pandas as pd
import numpy as np
```

```python
path_1 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/inventory.csv'
path_2 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/products.csv'
path_3 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/stores.csv'
path_4 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/sales.csv'
```

```python
inventory = pd.read_csv(path_1)
products = pd.read_csv(path_2)
stores = pd.read_csv(path_3)
sales = pd.read_csv(path_4)
```

```python
data1 = pd.merge(sales, stores, on='Store_ID')
data1
```

```python
data1.info()
```

```python
data2 = pd.merge(inventory, products, on='Product_ID')
data2
```

```python
data2.info()
```

```python
# data1

for dtype in ['int','object']:
    selected_dtype = data1.select_dtypes(include=[dtype])
    mean_usage_b = selected_dtype.memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage_b / 1024 ** 2
    print("Average memory usage for {} columns: {:03.2f} MB".format(dtype,mean_usage_mb))
```

```python
# data2

for dtype in ['int','object']:
    selected_dtype = data2.select_dtypes(include=[dtype])
    mean_usage_b = selected_dtype.memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage_b / 1024 ** 2
    print("Average memory usage for {} columns: {:03.2f} MB".format(dtype,mean_usage_mb))
```

```python
dataset = pd.merge(data1, data2, on='Product_ID',how='inner')
dataset
```

```python
dataset.info()
```

```python
dataset.to_csv('/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/maven_datatoys.csv', index=False, sep=';')
```
