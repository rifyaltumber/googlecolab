```python
from google.colab import drive

import pandas as pd
import numpy as np

drive.mount('/content/drive')
```

```python
path_1 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/inventory.csv'
path_2 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/products.csv'
path_3 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/stores.csv'
path_4 = '/content/drive/MyDrive/Kumpulan Dataset/2021-Dataset /Maven/Mexico Toy Sales/sales.csv'
```

```python
inventory = pd.read_csv(path_1)
products = pd.read_csv(path_2)
stores = pd.read_csv(path_3)
sales = pd.read_csv(path_4)
```

```python
# Soal-1 & 2 (A)
# Which product categories drive the biggest profits? Is this the same across store locations?

data1 = pd.merge(products, sales, on='Product_ID')
data1
```

```python
# Soal-1 & 2 (B)
# Which product categories drive the biggest profits? Is this the same across store locations?

data2 = pd.merge(stores, data1, on='Store_ID')
data2
```
